```python
import numpy as np
import matplotlib.pyplot as plt
import os

def plot_training_loss(output_dir):
    """
    绘制训练损失曲线
    
    Args:
        output_dir (str): 训练输出目录，包含损失数据文件
    """
    # 加载epoch损失数据
    epoch_loss_path = os.path.join(output_dir, "epoch_losses_data.npy")
    
    try:
        # 加载epoch损失
        epoch_data = np.load(epoch_loss_path, allow_pickle=True).item()
        epoch_losses = epoch_data['epoch_losses'][:45]
        epochs = list(range(1, len(epoch_losses) + 1))
        
        print(f"成功加载 {len(epoch_losses)} 个epoch的损失数据")
        
    except Exception as e:
        print(f"加载损失数据失败: {e}")
        return
    
    # # 创建图表
    # plt.figure(figsize=(12, 8))
    
    # # 绘制epoch损失曲线
    # plt.subplot(2, 1, 1)
    # plt.plot(epochs, epoch_losses, 'b-o', linewidth=2, markersize=4, label='Epoch Loss')
    plt.figure(figsize=(30, 10)) 
    plt.plot(epochs, epoch_losses, marker='o', linestyle='-', color='b', label='Epoch Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training Loss per Epoch')
    plt.grid(True, alpha=0.3)
    plt.legend()

if __name__ == "__main__":
    output_dir = "../megadna_trained_model"
    plot_training_loss(output_dir)
```
